````python
# A binary tree node
class Node:
    # Constructor to create a new node
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
````

````python
# Compute the "maxDepth" of a tree → the number of nodes
# along the longest path from the root node down to the farthest leaf node
def maxDepth(node):
    if node is None:
        return -1 ;

    else :

        # Compute the depth of each subtree
        lDepth = maxDepth(node.left)
        rDepth = maxDepth(node.right)

        # Use the larger one
        if (lDepth > rDepth):
            return lDepth+1
        else:
            return rDepth+1
````

````python
# Driver program to test above function
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
````

````python
print ("Height of tree is %d" %(maxDepth(root)))

# This code is contributed by Amit Srivastav
````

## Logika Program Graph Height (maxDepth)

### 1. Class `Node`

Merepresentasikan satu simpul pada binary tree. Setiap node menyimpan tiga hal: `data` (nilai), `left` (anak kiri), dan `right` (anak kanan). Nilai awal `left` dan `right` adalah `None` (belum punya anak).

### 2. Fungsi `maxDepth(node)` — Rekursif

Fungsi ini menghitung **kedalaman maksimum** pohon, yaitu panjang jalur terpanjang dari root ke leaf.

**Base case** — jika `node is None`, kembalikan `-1`. Ini terjadi saat kita melewati leaf node (anak dari leaf adalah `None`).

**Rekursif case:**
1. Hitung kedalaman subtree kiri → `lDepth = maxDepth(node.left)`
2. Hitung kedalaman subtree kanan → `rDepth = maxDepth(node.right)`
3. Ambil yang lebih besar, tambah 1 → `max(lDepth, rDepth) + 1`

### 3. Struktur Pohon yang Diuji

```
        1          ← depth 0
       / \
      2   3        ← depth 1
     / \
    4   5          ← depth 2
```

### 4. Jejak Rekursi

| Panggilan | lDepth | rDepth | Return |
|---|---|---|---|
| `maxDepth(4)` | -1 | -1 | 0 |
| `maxDepth(5)` | -1 | -1 | 0 |
| `maxDepth(3)` | -1 | -1 | 0 |
| `maxDepth(2)` | 0 | 0 | 1 |
| `maxDepth(1)` | 1 | 0 | **2** |

### 5. Hasil

```
Height of tree is 2
```

> **Mengapa return `-1` bukan `0`?** Karena kita menghitung **edge** (jumlah tingkat), bukan node. Leaf node itu sendiri bernilai 0, dan `None` (di bawah leaf) bernilai -1, sehingga saat ditambah 1 hasilnya tepat 0 untuk leaf.
