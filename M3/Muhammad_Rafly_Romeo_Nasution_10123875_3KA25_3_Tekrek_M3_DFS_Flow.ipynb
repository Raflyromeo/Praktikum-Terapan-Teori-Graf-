## Studi Kasus

Tujuan dari program ini adalah melakukan pencarian jalur dari kota **Arad** ke **Bucharest** menggunakan metode **Depth-First Search (DFS)** pada sebuah graf yang merepresentasikan peta jalan antar kota.

## Langkah 1 — Mendefinisikan Graf

Pertama-tama, dibuatlah sebuah struktur data berupa **dictionary** (kamus) Python untuk merepresentasikan graf. Setiap kunci (*key*) adalah nama sebuah simpul/kota, dan nilainya (*value*) adalah daftar kota-kota yang bertetangga langsung (terhubung langsung).

```python
graph = {
  'A'  : ['Z', 'S', 'T'],
  'Z'  : ['O'],
  'O'  : ['S1'],
  'S1' : ['F', 'R'],
  'F'  : ['B'],
  'B'  : [],
  'R'  : [],
  'S'  : ['O1', 'F1', 'R1'],
  'O1' : [],
  'F1' : ['B1'],
  'B1' : [],
  'R1' : ['C', 'P'],
  'C'  : [],
  'P'  : [],
  'T'  : ['L'],
  'L'  : ['M'],
  'M'  : ['D'],
  'D'  : []
}
```

**Artinya:**
- Kota `A` terhubung ke `Z`, `S`, dan `T`
- Kota `Z` hanya terhubung ke `O`
- Kota `B`, `R`, `O1`, `B1`, `C`, `P`, dan `D` adalah simpul akhir (tidak punya tetangga), artinya ini adalah **daun (leaf node)** dalam graf

---



## Langkah 2 — Menyiapkan Pelacak Simpul yang Sudah Dikunjungi

Sebelum penelusuran dimulai, dibuatlah sebuah **set** kosong bernama `visited`. Fungsinya adalah mencatat simpul mana saja yang sudah pernah dikunjungi selama proses DFS berlangsung.

```python
visited = set()
```

Kenapa pakai `set`? Karena `set` secara otomatis menghindari duplikat — setiap simpul hanya akan muncul sekali meski didaftarkan berkali-kali.

---



## Langkah 3 — Membuat Fungsi DFS (Rekursif)

Inti dari program ini ada di fungsi `dfs()` yang bersifat **rekursif** (memanggil dirinya sendiri).

```python
def dfs(visited, graph, node):
    if node not in visited:
        print(node)
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)
```

**Logika di dalam fungsi ini:**

1. **Cek apakah simpul sudah dikunjungi** — Jika `node` belum ada di dalam `visited`, maka lanjut ke langkah berikutnya. Jika sudah ada, lewati saja (tidak diproses ulang).

2. **Cetak nama simpul** — Simpul yang baru dikunjungi ditampilkan ke layar sebagai bukti bahwa penelusuran sedang melewati simpul tersebut.

3. **Tandai sebagai sudah dikunjungi** — Simpul dimasukkan ke dalam `visited` agar tidak dikunjungi lagi.

4. **Kunjungi tetangga satu per satu** — Untuk setiap tetangga (`neighbour`) dari simpul saat ini, fungsi `dfs()` dipanggil lagi secara rekursif. Ini yang membuat DFS menelusuri **sedalam mungkin** ke satu cabang sebelum berpindah ke cabang lain.

---



## Langkah 4 — Menjalankan Penelusuran

Penelusuran dimulai dari simpul **`'A'`** (merepresentasikan kota Arad):

